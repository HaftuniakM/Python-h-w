```python
# The least profitable 10 movies in the dataset
df_1 = df_data.sort_values(by = 'Profitability', ascending = True).head(10)

#The most profitable 10 movies in the dataset
df_2 = df_data.sort_values(by = 'Profitability', ascending = False).head(10)

display(df_1, df_2)
```

```python
# If we need one common DataFrame, we can concat 2 DataFrames:
df_all_data = pd.concat([df_1, df_2], axis = 0)
df_all_data
```

```python
# 7. Значение «Жанр» временами кажется непоследовательным; попробуйте найти эти несоответствия и исправить их.
```

```python
# Let's check what inconsistencies we have:
df_data.Genre.unique()

# inconsistencie_1: 'Comedy' != 'Comdy' = !'comedy'
# inconsistencie_2: 'Romance' != 'Romence' = !'romance'
```

```python
# Let's eliminate inconsistencie_1
df_data['Genre'] = df_data['Genre'].str.replace('comedy', 'Comedy', regex=False)
df_data['Genre'] = df_data['Genre'].str.replace('Comdy', 'Comedy', regex=False)

# Let's eliminate inconsistencie_2
df_data['Genre'] = df_data['Genre'].str.replace('Romence', 'Romance', regex=False)
df_data['Genre'] = df_data['Genre'].str.replace('romance', 'Romance', regex=False)
```

```python
# Let's check if everything is ok:
df_data.Genre.unique()
```

```python
# 8. Сохраните (в новый файл CSV) 10 лучших комедий по количеству зрителей; 
# покажите только название фильма, год и студию.
```

```python
# Top 10 comedies by Audience score %:
top_com = df_data[df_data.Genre.isin(['Comedy'])].sort_values(by = 'Audience score %', ascending = False).head(10)

# Showing only Movie title, Year and Lead Studio:
final_top_com = top_com.loc[:, ('Film', 'Year', 'Lead Studio')]
```

```python
# Saving (to a new CSV file):
final_top_com.to_csv('C:\\VS_Code\\final_top_com.csv', index=False, sep=';')
```

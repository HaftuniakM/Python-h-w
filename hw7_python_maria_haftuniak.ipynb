```python
import pandas as pd
from mysql.connector import connect
```

```python
# Python homework 7
# 1. Прочтите этот набор данных по пауэрлифтингу в pandas DataFrame.
file = 'https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2019/2019-10-08/ipf_lifts.csv'
df = pd.read_csv(file)
df.head()
```

```python
# 2. Найдите рекорды каждого пола и каждого подразделения в каждом из 3 упражнений (жим лежа, приседания и становая тяга).

# Solution
# Вирішила виконати завдання альтернативним способом з допомогою методів pandas

# Виділили підрозділи, які є в датасеті - 12 шт. 
# (Open, Juniors, Masters 1, Masters 2, Sub-Juniors, Masters 3, Masters 4, Light, Heavy, Middle, Super, SuperHeavy)
df.division.value_counts() 
```

```python
# За допомогою методів contains фільтрую підрозділ та стать, 
# За допомогою sort_values обираю вправу та сортую результати у порядку спадання, щоб отримати рекорди
df[df.division.str.contains('Open', na=False)][df.sex.str.contains('M')].sort_values(by = 'best3squat_kg', ascending = False).head(3)
```

```python
df[df.division.str.contains('Open', na=False)][df.sex.str.contains('M')].sort_values(by = 'best3bench_kg', ascending = False).head(3)
```

```python
df[df.division.str.contains('Open', na=False)][df.sex.str.contains('M')].sort_values(by = 'best3deadlift_kg', ascending = False).head(3)
```

```python
# 3. Посчитайте количество побед каждого участника, принимая за победу 1-е место. 
# Сохраните результаты во вторичном DataFrame.

# Solution
# Вибираю дані по учасникам, які зайняли 1 місце:
first_place = df.loc[df['place'] == '1']

# Рахую кількість перемог по кожному учаснику:
result = pd.DataFrame(first_place.groupby(['name']).agg(num_wins = ('place', 'count')).sort_values('num_wins', ascending=False))
result.head(10)
```

```python
# 3 - Перевіряю отримані результати з допомогою підключення до даного файлу в MySQL
# Результати співпадають
conn = connect(host = '127.0.0.1',
               port = 3306,
               user = 'root',
               password = 'root',
               database = 'test_db_da_1')
cursor = conn.cursor(buffered=True)

cursor.execute("""SELECT `name`, COUNT(place) AS `num_wins`
                    FROM ipf_lifts 
                    WHERE place = '1'
                GROUP BY `name`
                ORDER BY `num_wins` DESC""")
data = pd.DataFrame(cursor.fetchall(), columns=['name', 'num_wins'])
data.head(10)
```

```python
# 4. Используйте DataFrame из упражнения 3, 
# чтобы показать для каждой комбинации подразделения и пола, участника с наибольшим количеством побед
# среди участников, когда-либо участвовавших в этом дивизионе.

# Solution
# Виконала завдання з допомогою підключення до даного файлу в MySQL:
cursor.execute("""SELECT *
FROM (SELECT sex, division, `name`, COUNT(place) as `Quantity of wins`
		FROM ipf_lifts
	   WHERE place = '1'
	GROUP BY sex, division, `name`
	ORDER BY `Quantity of wins` DESC) AS table1
GROUP BY sex, division""")
data = pd.DataFrame(cursor.fetchall())
data
```
